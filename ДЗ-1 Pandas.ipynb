```python
import numpy as np
import pandas as pd
```

## Задание 1

```python
authors = pd.DataFrame({'author_id': [1, 2, 3], 
                        'author_name': ['Тургенев', 'Чехов', 'Островский']}, 
                       columns = ['author_id', 'author_name'])
authors
```

```python
book = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3],
                    'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
                    'price':[450, 300, 350, 500, 450, 370, 290]},
                   columns = ['author_id', 'book_title', 'price'])
book
```

## Задание 2

```python
authors_price = pd.merge(authors, book, on = 'author_id')
authors_price
```

## Задание 3

```python
top5 = authors_price.sort_values(by = 'price').tail()
top5
```

## Задание 4

```python
a = authors_price.groupby('author_name')['price'].max()
b = authors_price.groupby('author_name')['price'].min()
c = authors_price.groupby('author_name')['price'].mean()
authors_stat = pd.DataFrame({'min_price': a,
                            'max_price': b,
                            'mean_price': c})
authors_stat   
```

## Задание 5

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
book_info = pd.pivot_table(authors_price, values = 'price', 
                           index = ['author_id', 'author_name'], columns = ['cover'], aggfunc=np.sum)
book_info['мягкая'].fillna(0, inplace = True)
book_info['твердая'].fillna(0, inplace = True)
book_info
```
